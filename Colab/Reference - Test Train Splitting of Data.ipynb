```python
# from https://www.youtube.com/watch?v=xgDs0scjuuQ
# realization = test_train_split requires inputs of ARRAYS not dataframes

import numpy as np
import pandas as pd

data_url = "https://raw.githubusercontent.com/harshitroy2605/splitting-dataset-into-test-and-train/master/cars.csv"
df=pd.read_csv(data_url,delimiter=';')

df

```

```python
df.shape
```

```python
column_names= df.columns
column_names
```

```python

features = column_names[1:]
label = column_names[0]
display(features, label)
```

```python
# in amii this is often referred to as the features
#independent=df.iloc[1:,1:0].values
independent = df[features].values
independent

```

```python
# in amii this is often referred to as the label
#dependent=df.iloc[1:,9:].values
dependent = df[label].values
dependent
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.2,random_state=0)

x_train

```
